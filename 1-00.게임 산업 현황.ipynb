```python
import pandas as pd
```

```python
data = pd.read_csv('./data/게임산업__총괄_20240206144524.csv',encoding='cp949')
data
```

```python
data = data[['시점','항목','매출액']]
data_total = data[data['항목']=='총계']
data_total = data_total.drop('항목',axis=1)
data_total = data_total.reset_index(drop=True)
data_total
```

```python
# 2022년 국내 게임산업 매출 : 21184768(백만원) 출처 : 한국콘텐츠진흥원, https://www.gamemeca.com/view.php?gid=1738185
```

```python
data_total.info()
```

```python
new_2021 = {'시점':'2022','매출액':'21184768'}
data_total.loc[len(data_total)] = new_2021
data_total['매출액'] = data_total['매출액'].astype('float')/100     #억 단위로 변환
data_total
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# 그래프 수치랑 모양 이상함
plt.rcParams['font.family']='paybooc'
plt.figure(figsize=(4,4))
sns.barplot(x='시점',y='매출액',data=data_total,color='pink')

plt.ylabel('매출액 (단위:억)')
plt.xlabel('연도')
plt.xticks(rotation=45)
plt.title('2012-2022 국내 게임 산업 매출액')
```
