```python
# def get_title(soup):
#     title = soup.find("span", class_="mw-page-title-main")
#     return title.text.strip()

def get_instuction(soup):
    content = soup.find(class_="mw-parser-output")
    content.text
    paragraphs = content.find_all(['p', 'h2'])
    
    introduction = []
    for item in paragraphs:
        introduction.append(item.text.strip())
    return " ".join(introduction)
```

```python
import requests
from bs4 import BeautifulSoup
import time

for index, link in enumerate(full_list["zeldadungeon_links"]):
    res = requests.get(link)
    html = res.text
    soup = BeautifulSoup(html,"html.parser")
    instruction = get_instuction(soup)
    full_list.at[index,"instruction"] = instruction
    time.sleep(2)
full_list["instruction"]    
```

### Shrine coodinater
- https://www.eurogamer.net/shrine-locations-maps-zelda-tears-of-the-kingdom-9343

```python
url = "https://www.eurogamer.net/shrine-locations-maps-zelda-tears-of-the-kingdom-9343"
res = requests.get(url)
```

```python
html = res.text
coodinater_soup = BeautifulSoup(html,"html.parser")
content_all = coodinater_soup.find(class_="article_body")
content = content_all.find_all("li")
content
```

```python
coordinate_list = []
for item in content:
    text = item.text.strip()
    if "(" in text:
        coodinater = text.split(")")[0]
        if "." in coodinater:
            final = coodinater.split(".")[1]
            coordinate_list.append(final)
        else:
            
            coordinate_list.append(coodinater)
        
len(coordinate_list)
```

```python
coordinate_df = pd.DataFrame()
for item in coordinate_list[:152]:
    shrine_name = item.split("(")[0].strip()
    coordinate_df.at[shrine_name, "coordinates"] = item.split("(")[1].replace(")", "")
    
coordinate_df.reset_index(inplace=True)
coordinate_df.rename(columns={"index": "Shrines", "coordinates": "Coordinates"}, inplace=True)
coordinate_df
```

```python
coordinate_df.to_csv("./shrines location",index=False)
```

```python
df1 = pd.read_html(str(table))[0]
df1.drop(df1.columns[2], axis=1, inplace=True)
df1.columns = [["categorize","Simosiwak Shrine"]]
df1
```

```python
import time

merged_table = df1
for link in full_list["zeldadungeon_links"]:
    res = requests.get(link)
    html=res.text
    soup=BeautifulSoup(html,"html.parser")
    table = soup.find("table",class_="infotable")
    title = soup.find("span", class_="mw-page-title-main")
    df = pd.read_html(str(table))[0]
    df.drop(df.columns[2], axis=1, inplace=True)
    df.columns = [["categorize",title.text]]
    merged_table = pd.merge(merged_table, df, how="outer")
    time.sleep(2)
merged_table
```
